# WhiteChapel Song's Lyrics Exploration

```python
import pylyrics3
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import operator
```

```python
whitechapel_lyrics = pylyrics3.get_artist_lyrics('System of a Down')
lyrics = list((whitechapel_lyrics.values()))
data = [lyric.lower().rstrip() for lyric in lyrics]
```

```python
data = [lyric.lower().rstrip() for lyric in lyrics]
```

```python
tokenizer = RegexpTokenizer(r'\w+')
words = []
for word in data:
    words.append(tokenizer.tokenize(word))

flat_list = []
for items in words:
    for item in items:
        flat_list.append(item)
```

```python
words = ' '.join(flat_list)
print(words)
```

## Calculating the word frequency

```python
stop_words = set(stopwords.words('english')) 
word_tokens = word_tokenize(words)
filtered_sentence = [w for w in word_tokens if not w in stop_words]
```

```python
dicts = {}
for word in filtered_sentence:
    if word not in dicts:
        dicts[word] = 1
    else:
        dicts[word] += 1

sorted_dict = sorted(dicts.items(), key=operator.itemgetter(1),reverse = True)
print(sorted_dict [:10])
```

##  Creating word cloud

```python
wc = WordCloud(background_color="black",width = 1920,height = 720, max_words=200).generate(words)
plt.figure(figsize = (15,10), dpi = 500)
plt.imshow(wc)
plt.axis('off')
plt.show()
plt.savefig('whitechapel.png')
```
